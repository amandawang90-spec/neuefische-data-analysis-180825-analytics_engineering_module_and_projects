```python
# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
pg_api_key  =config['X-RapidAPI-Key']
```

```python
import pprint
```

```python
import requests

url = "https://meteostat.p.rapidapi.com/stations/monthly"

querystring = {"station":"10637","start":"2020-01-01","end":"2020-12-31"}

headers = {
	"x-rapidapi-key": pg_api_key, #use the value of env file
	"x-rapidapi-host": "meteostat.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

result = response.json()

pprint.pprint(result) 
```

```python
# If you see 200, the request succeeded ✅.
print(response.status_code)
```

```python
import requests

url = "https://meteostat.p.rapidapi.com/stations/monthly"

querystring = {"station":"10488","start":"2020-01-01","end":"2020-12-31"}

headers = {
	"x-rapidapi-key": pg_api_key,
	"x-rapidapi-host": "meteostat.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

result=response.json()

pprint.pprint(result)
```

```python
# If you see 200, the request succeeded ✅.
print(response.status_code)
```

```python
weather_monthly_df=result
```

```python
import pandas as pd

weather_monthly_df=pd.json_normalize(weather_monthly_df)
```

```python
# weather monthly raw data:
weather_monthly_df
```

```python
# try to flatten the data, meta data is not needed, only data is needed.
# 0 is because there is only one row in the table of the raw data.
# [0, 'data'] is helping me to extract the weather monthly data for the following station:
# "station":"10488","start":"2020-01-01","end":"2020-12-31"
weather_monthly_df.loc[0,'data']
```

```python
weather_monthly_df_station_10488=pd.json_normalize(weather_monthly_df.loc[0,'data'])
```

```python
weather_monthly_df_station_10488
```

```python
from sqlalchemy import create_engine, types
from sqlalchemy import text # to be able to pass string
```

```python
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'
```

```python
engine = create_engine(url, echo=False)
```

```python
engine.url # password is hidden
```

```python
my_schema = 's_jingwang' # update it to your schema

with engine.begin() as conn: 
    result = conn.execute(text(f'SET search_path TO {my_schema};'))
```

```python
weather_monthly_df_station_10488.to_sql('weather_monthly_df_station_10488', engine, if_exists='replace',index=True)
```
