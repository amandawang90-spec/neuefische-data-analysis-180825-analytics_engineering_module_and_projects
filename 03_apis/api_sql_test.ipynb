## write data from pandas to sql: two options

### Version 1: Using SET search_path

```python
from sqlalchemy import create_engine, types
from sqlalchemy import text # to be able to pass string

# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}' #the same like version 1

# let's switch the logging off again. .
engine = create_engine(url, echo=False) #the same like version 1

my_schema = pg_schema 

with engine.begin() as conn:
    conn.execute(text(f'SET search_path TO {my_schema};'))

weather_monthly_df_station_10488.to_sql('weather_monthly_df_station_10488_test_1', engine, if_exists='replace',index=True)
```

### Version 2: Using schema parameter in to_sql

```python
# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}' #the same like version 1

# let's switch the logging off again. .
engine = create_engine(url, echo=False) #the same like version 1


# writing dataframe to DB : Pandas Dataframe to DB Table in my own Schema
weather_monthly_df_station_10488.to_sql(name = 'weather_monthly_df_station_10488_test_2', 
                       con = engine, 
                       schema = pg_schema, # pandas is allowing to specify, in which schema the table shall be created
                       if_exists='replace', 
                       index=False
                      )
```

## read data from sql to pandas

### version 1: Using SET search_path/my_schema: my_schema = pg_schema is defined as a separate variable.

```python
# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
pg_api_key  =config['X-RapidAPI-Key']

# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'

# let's switch the logging off again. .
engine = create_engine(url, echo=False)

my_schema = pg_schema # update it to your schema

with engine.begin() as conn: 
    result = conn.execute(text(f'SET search_path TO {my_schema};'))

query_weather_monthly_df_station_10488_test_2 ='''
SELECT * FROM weather_monthly_df_station_10488_test_2
LIMIT 10
'''

weather_monthly_df_station_10488_test_2 = pd.read_sql(query_weather_monthly_df_station_10488_test_2, con=engine)
weather_monthly_df_station_10488_test_2.head()
```

### version 2: Using SET search_path/pg_schema: directly use pg_schema without defining my_schema

```python
import pandas as pd

# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
pg_api_key  =config['X-RapidAPI-Key']

# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'

# let's switch the logging off again. .
engine = create_engine(url, echo=False)

with engine.begin() as conn: 
    result = conn.execute(text(f'SET search_path TO {pg_schema};'))

query_weather_monthly_df_station_10488_test_2 =f'''
SELECT * FROM weather_monthly_df_station_10488_test_2
LIMIT 10;
'''
weather_monthly_df_station_10488_test_2 = pd.read_sql(query_weather_monthly_df_station_10488_test_2, con=engine)
weather_monthly_df_station_10488_test_2.head()
```

### version 3: don’t set the search path, but explicitly include the schema in every query.
### Example: df = pd.read_sql("SELECT * FROM my_schema.weather_data", con=engine)

```python
import pandas as pd

# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
pg_api_key  =config['X-RapidAPI-Key']

# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'

# let's switch the logging off again. .
engine = create_engine(url, echo=False)

query_weather_monthly_df_station_10488_test_2 =f'''
SELECT * FROM {pg_schema}.weather_monthly_df_station_10488_test_2
LIMIT 10;
'''
weather_monthly_df_station_10488_test_2 = pd.read_sql(query_weather_monthly_df_station_10488_test_2, con=engine)
weather_monthly_df_station_10488_test_2.head()
```

### Summary: where to put the schema connection

```python
# | Version | Schema Handling                              | Pros                          | Cons                               |
# | ------- | -------------------------------------------- | ----------------------------- | ---------------------------------- |
# | **1**   | `SET search_path` (via connection)           | Clean SQL; schema set once    | Requires session command each time |
# | **2**   | Same as #1 (slightly cleaner variable usage) | Same as above                 | Same as above                      |
# | **3**   | Schema included in SQL query                 | Portable; no need to set path | Longer SQL; schema repeated        |
```

## Summary: to_sql/pd.read.sql

```python
# Summary of df.to_sql/pd.read_sql
# | Goal                             | Correct function                         | Example                                                     |
# | -------------------------------- | ---------------------------------------- | ----------------------------------------------------------- |
# | Write a DataFrame → PostgreSQL   | `.to_sql()`                              | `df.to_sql(name='mytable', con=engine, schema='myschema')`  |
# | Read from PostgreSQL → DataFrame | `pd.read_sql()` or `pd.read_sql_table()` | `pd.read_sql('SELECT * FROM myschema.mytable', con=engine)` |
```

```python
# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'

# let's switch the logging off again. .
engine = create_engine(url, echo=False)

# writing dataframe to DB : Pandas Dataframe to DB Table in my own Schema

weather_monthly_df_station_10488_test_2.read_sql(
                       con = engine, 
                       schema = pg_schema, # pandas is allowing to specify, in which schema the table shall be created
                      )
```
