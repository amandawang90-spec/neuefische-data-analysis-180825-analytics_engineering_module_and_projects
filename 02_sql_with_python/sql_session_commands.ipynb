```python
# 1️⃣ What a session is
# A session in PostgreSQL is essentially a connection to the database.
# Each time you connect, PostgreSQL creates a new session.
# Any session-level settings (like timezone, search_path, or custom variables) last only for the duration of that session.
# Think of it like opening a web browser: settings you change in that browser window (session) only apply while it’s open. Close it, and they reset.
```

```python
# 2️⃣ What a session command is
# A session command is a SQL command that changes some behavior only for the current session.
# Common examples:
# SET search_path TO my_schema; → Changes the default schema for queries in this session.
# SET TIME ZONE 'UTC'; → Changes the timezone for timestamps in this session.
# SET work_mem = '64MB'; → Adjusts memory available for operations like sorting/joining for this session.
# These commands don’t change the database globally, only for the connection you’re using.
```

```python
# -- Connect to the database
# -- Then run:
# SET search_path TO my_schema;
# SET TIME ZONE 'UTC';

# -- Now run your queries without schema prefix
# SELECT * FROM weather_table;

# weather_table will automatically be searched in my_schema because of the session command.
# Once you disconnect, the search_path resets.
```
